```python
# vamos a sumar dos columnas: atak y sp_attack y se agrega a una nueva columna

df_pokedex['total_attack']=df_pokedex['attack']+df_pokedex['sp_attack']
df_pokedex.head()
```

```python
# sumar todos los valore snumericos y asignarlos a una nueva variable
df_pokedex['total_stats']=df_pokedex.get(['hp','attack','defense','sp_attack','sp_defense','speed']).sum(axis=1) # axis=1 para mostrar columnas
df_pokedex.head()
```

##Apply

```python
df_pokedex['generation'].unique()
```

```python
def generaciones(numero):
    if numero == 1:
        return 'Kanto'
    elif numero == 2:
        return 'Johto'
    elif numero == 3:
        return 'Hoenn'
    elif numero == 4:
        return 'Sinnoh'
    elif numero == 5:
        return 'Unova'
    elif numero == 6:
        return 'Kalos'
    elif numero == 7:
        return 'Alola'
```

```python
df_pokedex['region']=df_pokedex['generation'].apply(generaciones)
df_pokedex
```
